```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import numpy as np
```

```python
uk = pd.DataFrame(columns = ['Title', 'created_at', 'URL', 'text', 'About_text', 'About_URL'])
```

```python
titles = []#check
links = []#check
authors = [] # authors don't exist
dates = []#check
descs = []#check
abouts_text = []#check
abouts_URL = []#check


for i in range(1,35):
  my_url = 'https://www.news247.gr/ukraine?pages='+str(i)
  req = Request(my_url,headers={'User-Agent': 'Mozilla/5.0'})
  # print(my_url)
  raw_html = urlopen(req).read()
  news = BeautifulSoup(raw_html, "html.parser")


  titles_soup = news.find_all("h2", {"class": "article__title bold"})

  # links = news.find_all("h2", {"class": "article__title bold"})
  for title in titles_soup:
    links.append(title.find('a')['href'])
    titles.append((title.text).replace('\n',''))

  abouts_soup = news.find_all("div", {"class": "article__summary"})
  for about in abouts_soup:
    info = (about.text).splitlines()
    # print(info)
    abouts_text.append(info[1])

    descs.append(info[2])

    dates.append(info[3])

    abouts_URL.append(about.find('a')['href'])

print(len(titles))
print(len(links))
print(len(dates))
print(len(descs))
print(len(abouts_text))
print(len(abouts_URL))
```

```python
# uk = pd.DataFrame(columns = ['Title', 'created_at', 'URL', 'text', 'About_text', 'About_URL'])

for i in range(0, len(titles)):
  uk = uk.append({'Title' : titles[i], 'created_at' : dates[i], 'URL' : links[i], 'text' : descs[i], 'About_text' : abouts_text[i], 'About_URL' : abouts_URL[i]}, ignore_index=True)
```

```python
uk_doc['Year'] = uk_doc.created_at.apply(lambda dt: dt.split(' ')[2])
uk_doc['Month'] = uk_doc.created_at.apply(lambda dt: dt.split(' ')[1])
uk_doc['Day'] = uk_doc.created_at.apply(lambda dt: dt.split(' ')[0])
uk_doc['Time'] = uk_doc.created_at.apply(lambda dt: dt.split(' ')[3])

months = {
    'Ιανουαρίου' : '01',
    'Φεβρουαρίου' : '02', 
    'Μαρτίου' : '03',
    'Απριλίου' : '04',
    'Μαΐου' : '05',
    'Ιουνίου': '06',
    'Ιουλίου' : '07',
    'Αυγούστου' : '08',
    'Σεπτεμβρίου' : '09',
    'Οκτωβρίου' : '10',
    'Νοεμβρίου' : '11',
    'Δεκεμβρίου' : '12',
     
}

uk_doc['Date'] = uk_doc.apply(lambda row: f'{months[row.Month]}/{row.Day}/{row.Year} {row.Time}', axis='columns',result_type='expand')
uk_doc['Date'] = pd.to_datetime(uk_doc['Date'])


uk_doc
```
